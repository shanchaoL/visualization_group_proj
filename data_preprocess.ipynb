```python
import numpy as np
import pandas as pd
import random
```

```python
DF = pd.read_csv('all_games.csv', encoding = 'UTF-8')
DF["release_year"] = DF["release_date"].str.slice(-4)
DF["platform"] = DF["platform"].str.strip()
DF.loc[:, 'P'] = "Others"
DF.loc[(DF.platform == "PC"), 'P'] = "PC"
DF.loc[(DF.platform == "PlayStation 4"), 'P'] = "PlayStation 4"
DF.loc[(DF.platform == "Xbox 360"), 'P'] = "Xbox 360"
DF = DF[(DF.P != "Others") ]
DF['release_year'] = DF['release_year'].apply(pd.to_numeric)
year_value = dict(DF["release_year"].value_counts())

gamma = 10
m = max(year_value)
for i in year_value.keys():
    DF.loc[DF['release_year'] == i, 'meta_score'] += i + random.randrange(-0.5,0.5)


DF = DF[DF.meta_score > 50 ]
DF["user_review"] = DF["user_review"].str.strip()
DF.loc[(DF.user_review == "tbd"), 'user_review'] = "0"
DF['user_review'] = DF['user_review'].apply(pd.to_numeric)
for i in year_value.keys():
    DF.loc[DF['release_year'] == i, 'users_review'] += i + random.randrange(-0.5,0.5)
DF = DF[DF["user_review"] >= 5]
DF = DF[DF["release_year"]> 2005 ]

DF
```

```python
DF.to_csv("selected_games_1.csv")
```
